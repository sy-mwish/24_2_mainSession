## check_winnerí•¨ìˆ˜
- ê°€ë¡œ, ì„¸ë¡œ, ëŒ€ê°ì„ ì„ í™•ì¸
- ìŠ¹ì or ë¬´ìŠ¹ë¶€ ë°˜í™˜

```python
class TictactoeState:
    def __init__(self):
        # 3x3 ë³´ë“œ ì´ˆê¸°í™” (0: ë¹ˆ ì¹¸, 1: í”Œë ˆì´ì–´ 1, 2: í”Œë ˆì´ì–´ 2)
        self.board = [[0 for _ in range(3)] for _ in range(3)]
        # í˜„ì¬ í”Œë ˆì´ì–´ (1: í”Œë ˆì´ì–´ 1, 2: í”Œë ˆì´ì–´ 2)
        self.current_player = 1

    def get_possible_moves(self):
        # ë‘˜ ìˆ˜ ìˆëŠ” ìœ„ì¹˜(ë¹ˆ ì¹¸ì˜ ì¢Œí‘œ ë¦¬ìŠ¤íŠ¸)ë¥¼ ë°˜í™˜
        moves = []
        for i in range(3):
            for j in range(3):
                if self.board[i][j] == 0:  # ë¹ˆ ì¹¸ì¼ ê²½ìš°
                    moves.append((i, j)) # ë¦¬ìŠ¤íŠ¸ì— ì¢Œí‘œ ì¶”ê°€
        return moves

    def check_winner(self):
        # í˜„ì¬ ë³´ë“œ ìƒíƒœì—ì„œ ìŠ¹ì, ë¬´ìŠ¹ë¶€ ì—¬ë¶€ë¥¼ í™•ì¸
        # ìŠ¹ì í™•ì¸: ê°€ë¡œ, ì„¸ë¡œ, ëŒ€ê°ì„  ì¤„ ê²€ì‚¬
        lines = [    # ê°€ë¡œ, ì„¸ë¡œ, ëŒ€ê°ì„ ì— í•´ë‹¹í•˜ëŠ” ëª¨ë“  ì¤„ì„ ìˆ˜ì§‘í•˜ì—¬ lines ë¦¬ìŠ¤íŠ¸ì— ì €ì¥
            # ê°€ë¡œ, ì„¸ë¡œ í™•ì¸
            [self.board[i] for i in range(3)],  # ê° í–‰
            [list(row[i] for row in self.board) for i in range(3)],  # ê° ì—´
            # ëŒ€ê°ì„  í™•ì¸
            [[self.board[i][i] for i in range(3)], [self.board[i][2 - i] for i in range(3)]]
        ]

        for group in lines:
            for line in group:
                if len(set(line)) == 1 and line[0] != 0:  # ì„¸ ì¹¸ì´ ë™ì¼í•˜ê³  ë¹ˆ ì¹¸ì´ ì•„ë‹ ë•Œ, ì§‘í•©(set)ìœ¼ë¡œ ë³€í™˜í•˜ì—¬ ì¤‘ë³µì„ ì œê±°
                    return line[0]  # ìŠ¹ìë¥¼ ë°˜í™˜ (1: í”Œë ˆì´ì–´ 1, 2: í”Œë ˆì´ì–´ 2)

        # ë¬´ìŠ¹ë¶€ í™•ì¸: ë¹ˆ ì¹¸ì´ ì—†ê³  ìŠ¹ìê°€ ì—†ì„ ë•Œ
        if not self.get_possible_moves():
            return "Draw"

        return None  # ì•„ì§ ìŠ¹ë¶€ê°€ ë‚˜ì§€ ì•ŠìŒ

    def make_move(self, row, col):
        # ì£¼ì–´ì§„ ìœ„ì¹˜ì— í˜„ì¬ í”Œë ˆì´ì–´ì˜ ìˆ˜ë¥¼ ë†“ëŠ”ë‹¤
        if self.board[row][col] == 0:  # ë¹ˆ ì¹¸ì¸ì§€ í™•ì¸
            self.board[row][col] = self.current_player # ìˆ˜ ë†“ê¸°(1 or 2)
            # ë‹¤ìŒ í„´ì„ ìœ„í•œ í”Œë ˆì´ì–´ ë³€ê²½
            self.current_player = 1 if self.current_player == 2 else 2
        else:
            raise ValueError("ë¹ˆ ì¹¸ì´ ì—†ë‹¤!")

    def render(self):
        # í˜„ì¬ ë³´ë“œ ìƒíƒœë¥¼ ì¶œë ¥
        symbols = {0: ' ', 1: 'X', 2: 'O'}
        for i in range(3):
            print(' | '.join(symbols[self.board[i][j]] for j in range(3)))
            if i < 2:
                print('---------')
        print()
```

# minimax ë° alpha_beta_pruning ë©”ì„œë“œ ì¶”ê°€

```python
class TictactoeState:
    def __init__(self):
        self.board = [[0 for _ in range(3)] for _ in range(3)]
        self.current_player = 1

    def get_possible_moves(self):
        moves = []
        for i in range(3):
            for j in range(3):
                if self.board[i][j] == 0:
                    moves.append((i, j))
        return moves

    def check_winner(self):
        lines = [
            [self.board[i] for i in range(3)],
            [list(row[i] for row in self.board) for i in range(3)],
            [[self.board[i][i] for i in range(3)], [self.board[i][2 - i] for i in range(3)]]
        ]

        for group in lines:
            for line in group:
                if len(set(line)) == 1 and line[0] != 0:
                    return line[0]

        if not self.get_possible_moves():
            return "Draw"

        return None

    def make_move(self, row, col):
        if self.board[row][col] == 0:
            self.board[row][col] = self.current_player
            self.current_player = 1 if self.current_player == 2 else 2
        else:
            raise ValueError("ë¹ˆ ì¹¸ì´ ì—†ë‹¤")

    def undo_move(self, row, col):
        # ì´ì „ì— ë‘” ìˆ˜ë¥¼ ì·¨ì†Œí•˜ê³  í”Œë ˆì´ì–´ ìˆœì„œë¥¼ ë°”ê¾¼ë‹¤
        self.board[row][col] = 0
        self.current_player = 1 if self.current_player == 2 else 2

    def render(self):
        symbols = {0: ' ', 1: 'X', 2: 'O'}
        for i in range(3):
            print(' | '.join(symbols[self.board[i][j]] for j in range(3)))
            if i < 2:
                print('---------')
        print()

    # Min-Max ì•Œê³ ë¦¬ì¦˜
    def minimax(self, maximizing):
        winner = self.check_winner()
        if winner == 1:
            return 1  # í”Œë ˆì´ì–´ 1 ìŠ¹ë¦¬
        elif winner == 2:
            return -1  # í”Œë ˆì´ì–´ 2 ìŠ¹ë¦¬
        elif winner == "Draw":
            return 0  # ë¬´ìŠ¹ë¶€

        if maximizing:
            max_eval = -float('inf')
            for move in self.get_possible_moves():
                self.make_move(move[0], move[1])
                eval = self.minimax(False)
                self.undo_move(move[0], move[1])
                max_eval = max(max_eval, eval)
            return max_eval
        else:
            min_eval = float('inf')
            for move in self.get_possible_moves():
                self.make_move(move[0], move[1])
                eval = self.minimax(True)
                self.undo_move(move[0], move[1])
                min_eval = min(min_eval, eval)
            return min_eval

    # ì•ŒíŒŒ-ë² íƒ€ ê°€ì§€ì¹˜ê¸°
    def alpha_beta_pruning(self, maximizing, alpha, beta):
        winner = self.check_winner()
        if winner == 1:
            return 1  # í”Œë ˆì´ì–´ 1 ìŠ¹ë¦¬
        elif winner == 2:
            return -1  # í”Œë ˆì´ì–´ 2 ìŠ¹ë¦¬
        elif winner == "Draw":
            return 0  # ë¬´ìŠ¹ë¶€

        if maximizing: # ìµœëŒ€í•œ ì ìˆ˜ ìµœëŒ€í™”
            max_eval = -float('inf')
            for move in self.get_possible_moves():
                self.make_move(move[0], move[1])
                eval = self.alpha_beta_pruning(False, alpha, beta)
                self.undo_move(move[0], move[1])
                max_eval = max(max_eval, eval)
                alpha = max(alpha, eval) # ë² íƒ€ ê°’ ê°±ì‹ 
                if beta <= alpha:
                    break  # ë² íƒ€ ê°€ì§€ì¹˜ê¸°
            return max_eval
        else:
            min_eval = float('inf') # ìµœëŒ€í•œ ì ìˆ˜ ìµœì†Œí™”
            for move in self.get_possible_moves():
                self.make_move(move[0], move[1])
                eval = self.alpha_beta_pruning(True, alpha, beta)
                self.undo_move(move[0], move[1])
                min_eval = min(min_eval, eval)
                beta = min(beta, eval) # ë² íƒ€ ê°’ ê°±ì‹ 
                if beta <= alpha:
                    break  # ì•ŒíŒŒ ê°€ì§€ì¹˜ê¸°
            return min_eval

    def best_move(self):
        # ë¯¸ë‹ˆë§¥ìŠ¤ í˜¹ì€ ì•ŒíŒŒ-ë² íƒ€ ê°€ì§€ì¹˜ê¸°ë¡œ ìµœì ì˜ ìˆ˜ë¥¼ ì°¾ëŠ”ë‹¤
        best_val = -float('inf')
        best_move = None
        for move in self.get_possible_moves():
            self.make_move(move[0], move[1])
            move_val = self.alpha_beta_pruning(False, -float('inf'), float('inf')) # ì•ŒíŒŒ-ë² íƒ€ ê°€ì§€ì¹˜ê¸°
            self.undo_move(move[0], move[1])
            if move_val > best_val:
                best_val = move_val
                best_move = move
        return best_move
```

```python
# ìµœì ì˜ ìˆ˜ë¥¼ ì°¾ëŠ”ë‹¤
move = state.best_move()
print("Best move:", move)

# ìµœì ì˜ ìˆ˜ë¥¼ ë‘”ë‹¤
if move:
    state.make_move(move[0], move[1])

state.render()
```

```python
def play_game():
    game = TictactoeState()

    print("í‹±íƒí†  ê²Œì„ì„ ì‹œì‘í•©ë‹ˆë‹¤!")
    game.render()

    while True:
        # í”Œë ˆì´ì–´ì˜ ì°¨ë¡€
        if game.current_player == 1:
            print("í”Œë ˆì´ì–´ 1 (X)ì˜ ì°¨ë¡€ì…ë‹ˆë‹¤.")
            try:
                row = int(input("í–‰ ë²ˆí˜¸ (0, 1, 2): "))
                col = int(input("ì—´ ë²ˆí˜¸ (0, 1, 2): "))
                game.make_move(row, col)
            except ValueError as e:
                print(e)
                continue
            except (IndexError, ValueError):
                print("ì˜¬ë°”ë¥¸ ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš” (0, 1, 2).")
                continue
        # AIì˜ ì°¨ë¡€
        else:
            print("í”Œë ˆì´ì–´ 2 (O, AI)ì˜ ì°¨ë¡€ì…ë‹ˆë‹¤.")
            move = game.best_move()
            if move:
                game.make_move(move[0], move[1])

        game.render()

        # ê²Œì„ ì¢…ë£Œ ì—¬ë¶€ í™•ì¸
        winner = game.check_winner()
        if winner is not None:
            if winner == 1:
                print("í”Œë ˆì´ì–´ 1 (X)ì´ ìŠ¹ë¦¬í–ˆìŠµë‹ˆë‹¤!")
            elif winner == 2:
                print("í”Œë ˆì´ì–´ 2 (O, AI)ê°€ ìŠ¹ë¦¬í–ˆìŠµë‹ˆë‹¤!")
            else:
                print("ë¬´ìŠ¹ë¶€ì…ë‹ˆë‹¤!")
            break

# ê²Œì„ ì‹œì‘
play_game()
```
