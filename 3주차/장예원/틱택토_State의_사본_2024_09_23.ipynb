**min-max algorithm / alpha-beta algorithm êµ¬í˜„(ì‹œë‹ˆì–´)**

### ğŸš¨ ë§Œì•½ ê°ì´ ì•ˆ ì˜¨ë‹¤ë©´ ìš´ì˜ì§„ì—ê²Œ ì—°ë½ì£¼ì‹œë©´ ì°¸ê³ í•  ë§Œí•œ ë ˆí¼ëŸ°ìŠ¤ë¥¼ ë³´ë‚´ë“œë¦¬ê² ìŠµë‹ˆë‹¤.
### ğŸš¨ ì°¸ê³ í•œ ë ˆí¼ëŸ°ìŠ¤ê°€ ìˆë‹¤ë©´, ì¶œì²˜ë¥¼ ë‚¨ê²¨ì£¼ì„¸ìš”.

```python
class TicTacToe:
    def __init__(self):
        self.board = [[0 for _ in range(3)] for _ in range(3)] # 3x3 ë³´ë“œ, 0ì€ ë¹ˆ ê³µê°„ì„ ì˜ë¯¸í•¨
        self.current_player = 1  # 1ë²ˆ í”Œë ˆì´ì–´ë¶€í„° ì‹œì‘

    def available_moves(self): #í˜„ì¬ ë³´ë“œì—ì„œ ë¹ˆ ê³µê°„ì„ ë°˜í™˜
        moves = [] #ë¹ˆ ê³µê°„ ë¦¬ìŠ¤íŠ¸
        for i in range(3):
            for j in range(3):
                if self.board[i][j] == 0:  # ë¹ˆ ê³µê°„ì´ë©´ ì¶”ê°€
                    moves.append((i, j))
        return moves

    def make_move(self, position): #í”Œë ˆì´ì–´ê°€ ìˆ˜ë¥¼ ë‘ëŠ” í•¨ìˆ˜   #position: ì‚¬ìš©ìê°€ ë‘ê³  ì‹¶ì€ ìœ„ì¹˜ë¥¼ (í–‰, ì—´) í˜•íƒœì˜ íŠœí”Œë¡œ ë°›ìŒ
        if self.board[position[0]][position[1]] == 0: #í˜„ì¬ ë³´ë“œì—ì„œ í”Œë ˆì´ì–´ê°€ ë‘ë ¤ëŠ” positionì´ ë¹ˆ ì¹¸ì¸ì§€ í™•ì¸í•˜ëŠ” ì¡°ê±´
            self.board[position[0]][position[1]] = self.current_player #í•´ë‹¹ ìœ„ì¹˜ì— í”Œë ˆì´ì–´ì˜ ìˆ˜ë¥¼ ë‘ 
            self.current_player = 2 if self.current_player == 1 else 1  # í”Œë ˆì´ì–´ ë³€ê²½ (í˜„ì¬ í”Œë ˆì´ì–´ê°€ 1ë²ˆì¸ ê²½ìš° 2ë²ˆìœ¼ë¡œ êµì²´)
        else: #í”Œë ˆì´ì–´ê°€ ì´ë¯¸ ìˆ˜ë¥¼ ë‘” ìœ„ì¹˜ë¥¼ ì„ íƒí–ˆì„ ê²½ìš° ì˜ˆì™¸ ë°œìƒ
            raise ValueError("ì´ë¯¸ ë‘” ìœ„ì¹˜ì…ë‹ˆë‹¤!")

    def check_winner(self): #ìŠ¹íŒ¨ í˜¹ì€ ë¬´ìŠ¹ë¶€ë¥¼ í™•ì¸í•˜ëŠ” í•¨ìˆ˜
        # ê°€ë¡œ, ì„¸ë¡œ, ëŒ€ê°ì„  ì²´í¬
        for i in range(3):
            # ê°€ë¡œ ìŠ¹ë¦¬
            if self.board[i][0] == self.board[i][1] == self.board[i][2] != 0: #ië²ˆì§¸ í–‰ì— ìˆëŠ” ì„¸ ê°œì˜ ì¹¸(ì—´ 0, 1, 2)ì´ ëª¨ë‘ ê°™ì€ ê°’, ê·¸ê²ƒë„ 0ì´ ì•„ë‹Œ ê°’ì¸ì§€ í™•ì¸
                return self.board[i][0] #í•´ë‹¹ ê°’ì„ ë°˜í™˜í•˜ì—¬ ìŠ¹ë¦¬ì ê²°ì •
            # ì„¸ë¡œ ìŠ¹ë¦¬
            if self.board[0][i] == self.board[1][i] == self.board[2][i] != 0: #ië²ˆì§¸ ì—´ì— ìˆëŠ” ì„¸ ê°œì˜ ì¹¸(í–‰ 0, 1, 2)ì´ ëª¨ë‘ ê°™ì€ ê°’ì„, ê·¸ê²ƒë„ 0ì´ ì•„ë‹Œ ê°’ì¸ì§€ í™•ì¸
                return self.board[0][i] #í•´ë‹¹ ê°’ì„ ë°˜í™˜í•˜ì—¬ ìŠ¹ë¦¬ì ê²°ì •

        # ëŒ€ê°ì„  ìŠ¹ë¦¬
        if self.board[0][0] == self.board[1][1] == self.board[2][2] != 0:
            return self.board[0][0] #í•´ë‹¹ ê°’ì„ ë°˜í™˜í•˜ì—¬ ìŠ¹ë¦¬ì ê²°ì •
        if self.board[0][2] == self.board[1][1] == self.board[2][0] != 0:
            return self.board[0][2] #í•´ë‹¹ ê°’ì„ ë°˜í™˜í•˜ì—¬ ìŠ¹ë¦¬ì ê²°ì •

        # ë¬´ìŠ¹ë¶€ í™•ì¸ (ë¹ˆ ì¹¸ì´ ì—†ìœ¼ë©´ ë¬´ìŠ¹ë¶€)
        if not self.available_moves():
            return "Draw"

        # ì•„ì§ ê²Œì„ì´ ëë‚˜ì§€ ì•ŠìŒ
        return None

    def render(self): #í˜„ì¬ ë³´ë“œ ìƒíƒœë¥¼ ì‹œê°ì ìœ¼ë¡œ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜
        symbols = {0: ".", 1: "X", 2: "O"} #ë¹ˆ ì¹¸ì€ .ìœ¼ë¡œ, 1ë²ˆ í”Œë ˆì´ì–´ëŠ” X, 2ë²ˆ í”Œë ˆì´ì–´ëŠ” Oë¡œ í‘œì‹œ
        for row in self.board:
            print(" ".join([symbols[cell] for cell in row])) #cell ê°’ì— ëŒ€ì‘í•˜ëŠ” ê¸°í˜¸ë¡œ ë³€í™˜í•˜ì—¬ ì¶œë ¥   #ë¦¬ìŠ¤íŠ¸ì— ìˆëŠ” ìš”ì†Œë“¤ì„ ë¬¸ìì—´ë¡œ ë³€í™˜í•˜ì—¬ ê³µë°± " "ì„ êµ¬ë¶„ìë¡œ ê²°í•©!
        print()
```

```python
# ì‚¬ìš© ì˜ˆì‹œ
game = TicTacToe()

# ì´ˆê¸° ìƒíƒœ ì¶œë ¥
game.render()

# í”Œë ˆì´ì–´ê°€ ìˆ˜ë¥¼ ë‘ê³  ìƒíƒœ ì¶œë ¥
game.make_move((0, 0))
game.render()

game.make_move((1, 1))
game.render()

# ê°€ëŠ¥í•œ ì´ë™ ìœ„ì¹˜ ì¶œë ¥
print("ê°€ëŠ¥í•œ ì´ë™ ìœ„ì¹˜:", game.available_moves())

# ìŠ¹ë¦¬ ì—¬ë¶€ í™•ì¸
winner = game.check_winner()
if winner:
    print(f"ìŠ¹ë¦¬ì: {winner}")
else:
    print("ì•„ì§ ìŠ¹ë¦¬ìê°€ ì—†ìŠµë‹ˆë‹¤.")
```
